```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

```

```python
df = pd.read_csv("cardA80.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.shape
```

```python
class_count = pd.value_counts(df['Class'])
sns.countplot(x="Class", data=df)
plt.title("Class Count")
plt.xlabel("Class")
plt.ylabel("Number of transactions")
class_count
```

# Feature Engineering

```python
df['Time_Day'] = df['Time'] / 3600 / 24
df['Time_Sin'] = np.sin(2 * np.pi * df['Time'] / 86400)
df['Time_Cos'] = np.cos(2 * np.pi * df['Time'] / 86400)
df['Amount_Day'] = df.groupby('Time_Day')['Amount'].transform('mean')
df
```

```python
from sklearn.preprocessing import StandardScaler
df['Amount_Day'] = StandardScaler().fit_transform(df['Amount'].values.reshape(-1, 1))
df['Amount'] = StandardScaler().fit_transform(df['Amount'].values.reshape(-1, 1))
df['Time_Day'] = StandardScaler().fit_transform(df['Time_Day'].values.reshape(-1, 1))
#df['Time'] = StandardScaler().fit_transform(df['Time'].values.reshape(-1, 1))
df
```

```python
df.dtypes
```

```python
X_train = df.drop(["Class","Time"], 1).values
Y_train = df["Class"].values
```

```python
from imblearn.over_sampling import SMOTE
oversampler = SMOTE(random_state=2)
X_train , Y_train = oversampler.fit_resample(X_train,Y_train)
```

```python
sns.countplot(x=Y_train)
print(X_train.shape,Y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators= 300,max_depth=15, criterion = 'entropy', random_state = 42, n_jobs = -1)
rf.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix,roc_auc_score,mean_squared_error
y_pred = rf.predict(X_train)
mse = mean_squared_error(Y_train, y_pred)
print('Mean squared error on test set:', mse)
```

```python
y_pred
```

```python
rf_accuracy = accuracy_score(Y_train, y_pred)
rf_recall = recall_score(Y_train, y_pred)
rf_cm = confusion_matrix(Y_train,y_pred)
rf_auc = roc_auc_score(Y_train, y_pred)

print("Accuracy: {:.5%}".format(rf_accuracy))
print("Recall: {:.5%}".format(rf_recall))
print("ROC AUC: {:.5%}".format(rf_auc))

rf_cm = pd.DataFrame(rf_cm, ['Actual Normal','Actual Fraud'],['Prediction Normal','Prediction Fraud'])
plt.figure(figsize = (8,4))
sns.set(font_scale=1.4)
sns.heatmap(rf_cm, annot=True,annot_kws={"size": 16},fmt='g',cmap="YlOrBr_r")
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
# Compute the ROC curve and the AUC
fpr, tpr, thresholds = roc_curve(Y_train, y_pred)
roc_auc = auc(fpr, tpr)
# Plot the ROC curve
plt.plot(fpr, tpr, lw=1, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], '--', color='gray', lw=1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
